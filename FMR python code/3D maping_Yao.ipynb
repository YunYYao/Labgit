```python
### for plot the multi-curves X-Y-Z 3D mapping  ###

#load the data into dataframe

def loadtxt_DF(filepath):
    dataload=pd.read_csv(filepath,index_col=0)
    dataload.reindex(index=[float(i) for i in dataload.index])
    data_DF=dataload.sort_index()
    
    return data_DF

 
#data_DF=sqa.fmr_load_data(data)
####
def mesh_DF_3D(data,datatype='3d'):
    if datatype=='3d':
        data=data-data.mean()
        
        
    X=np.array(data.index)
    Y=np.array([float(i) for i in data.columns])
    Z=data.values-0.0
    X,Y = np.meshgrid(Y,X)
    
    fig=plt.figure()
    ax=fig.gca()
    surf=ax.pcolormesh(X,Y,Z,vmin=Z.min(),vmax=Z.max(),linewidth=0,cmap=cm.rainbow_r)   ### gist_earth, ocean,gnuplot2,
    fig.colorbar(surf,shrink=0.5,aspect=10)
    plt.ylabel('Isd (uA)',fontsize=24)
    plt.yticks(fontsize=24)
    plt.xlabel('H (Oe)',fontsize=24)
    plt.xticks(fontsize=24)
    
    fig.show()


#data_filename='E:\\project\\S-F-S\\Nb-Py-Nb\\Josephson current detection\\Nb(100)-Py(5)-Nb(100)\\20200725\\H dep JJ\\results\\-H\\-H 4K.txt'
data_up_filename='E:\\project\\S-F-S\\Nb-Py-Nb\\Josephson current detection\\Nb(100)-Py(5)-Nb(100)\\20200725\\H dep JJ\\results\\4K.txt'
data_down_filename='E:\\project\\S-F-S\\Nb-Py-Nb\\Josephson current detection\\Nb(100)-Py(5)-Nb(100)\\20200725\\down 4k.txt'


#data_down=loadtxt_DF(data_down_filename)
#data_up=loadtxt_DF(data_up_filename)
data=loadtxt_DF(data_up_filename)

mapping=mesh_DF_3D(data)
```
